```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go


graph_template = "plotly_white"
print("Kütüphaneler başarıyla projenize dahil edildi!")
```

Bu çalışma, tekstil sektöründe faaliyet gösteren tam entegre Sanko Tekstil A.Ş.'nin çevresel ve iklimsel performans verileri ile bu verilerin oluşturduğu finansal riskleri analiz etmek amacıyla hazırlanmıştır. Analiz kapsamında, şirketin sera gazı emisyon dağılımları ve maruz kaldığı fiziksel/geçiş riskleri incelenmiştir.

```python
#(GRI 305)
emisyon_data = pd.DataFrame({
    'Kapsam': ['Kapsam 1 (Doğrudan Emisyonlar)', 'Kapsam 2 (Dolaylı - Elektrik)', 'Kapsam 3 (Değer Zinciri)'],
    'Emisyon (Ton CO2e)': [41000, 67000, 432000]
})

# Toplam emisyon miktarını otomatik hesaplayalım
toplam_emisyon = emisyon_data['Emisyon (Ton CO2e)'].sum()

# Plotly ile interaktif pasta grafiği çizimi
fig_emisyon = px.pie(
    emisyon_data, 
    values='Emisyon (Ton CO2e)', 
    names='Kapsam', 
    title=f'Sanko Tekstil Kapsam Bazlı Emisyon Dağılımı (Toplam: {toplam_emisyon:,} Ton CO2e)',
    color_discrete_sequence=px.colors.sequential.RdBu,
    template=graph_template
)

# Grafik üzeri metin ayarları
fig_emisyon.update_traces(textposition='inside', textinfo='percent+label')
fig_emisyon.show()
```

**GRI 305 Emisyon Analiz Notu:**  
Yapılan analiz sonucunda Sanko Tekstil'in toplam emisyon yükünün **%80'inden fazlasının (432.000 ton CO2e) Kapsam 3**, yani değer zinciri kaynaklı olduğu görülmektedir. Bu durum, şirketin dekarbonizasyon stratejisinde sadece kendi fabrikalarına değil, ham madde tedarikçilerine ve lojistik ağına da odaklanması gerektiğini net bir şekilde ortaya koymaktadır. 2025 sonu için belirlenen **%5 yoğunluk iyileştirmesi hedefi**, geçiş risklerini azaltmada kritik bir eşiktir.

```python
risk_data = pd.DataFrame({
    'Risk Kategorisi': ['Fiziksel', 'Fiziksel', 'Geçiş', 'Geçiş'],
    'Risk Faktörü': ['Su Kıtlığı / Kuraklık', 'Aşırı Sıcaklıklar', 'Sınırda Karbon (SKDM)', 'Müşteri Talebi Kayması'],
    'Risk Skoru (0-100)': [82, 65, 90, 74], # Müşteri talebi riski 74 olarak güncellendi
    'Uyum Stratejisi': [
        'Su geri kazanımını %32\'den %50\'ye çıkarma ve ileri arıtmaya yatırım.',
        'Isı yalıtımı, enerji verimli soğutma sistemleri ve çalışma saati düzenlemesi.',
        'Kapsam 2 emisyonlarını düşürmek için %24 yenilenebilir payını artırmak.',
        'Üretimin %100 sertifikalı liflere (GOTS/OCS/Geri Dönüşümlü) kaydırılması.'
    ]
})

fig_risk = px.bar(
    risk_data, 
    x='Risk Faktörü', 
    y='Risk Skoru (0-100)', 
    color='Risk Kategorisi',
    text='Risk Skoru (0-100)',
    title='Sanko Tekstil İklim ve Geçiş Riskleri Önceliklendirme Matrisi',
    color_discrete_sequence=['#e74c3c', '#3498db'],
    template=graph_template
)
fig_risk.update_layout(yaxis=dict(range=[0, 100]), xaxis_title="Risk Faktörleri", yaxis_title="Risk Skoru (0-100)")
fig_risk.show()

risk_data[['Risk Kategorisi', 'Risk Faktörü', 'Risk Skoru (0-100)', 'Uyum Stratejisi']]
```

## Stratejik Sonuç ve Gelecek Yol Haritası (2050 Vizyonu)

Yapılan iklim riski modellemesi ve önceliklendirme analizi sonucunda, Sanko Tekstil için finansal açıdan en kritik iki başlığın **AB Sınırda Karbon Düzenleme Mekanizması (SKDM - Skor: 90)** ve Gaziantep bölgesini doğrudan etkileyen **Bölgesel Su Kıtlığı (Skor: 82)** olduğu tespit edilmiştir.

Şirketin pazar payını koruması ve operasyonel sürekliliğini sağlaması adına geliştirilen aksiyon planı:
1. **Enerji ve Karbon Odaklılık:** %24 olan yenilenebilir enerji payının hızla artırılarak SKDM maliyet risklerinin absorbe edilmesi.
2. **Su Güvenliği:** Su kıtlığı riskine karşı mevcut %32 olan su geri kazanım kapasitesinin yeni arıtma teknolojileriyle %50+ seviyesine çıkarılması.
3. **Tedarik Zinciri Yönetimi (GRI 204):** Değer zinciri risklerini (Kapsam 3) azaltmak adına, tedarikçi denetim ve uyum oranının mevcut %68 seviyesinden %80'in üzerine taşınması.
